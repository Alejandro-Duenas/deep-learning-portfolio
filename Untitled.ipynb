## 1. Introduction
With the development of mahcine learning and the increasing disponibility of free high frequency financial data online, the oportunities to implement this techniques to forecast the behaviour and manage assets arise.

Also I use the [`quandl`](https://github.com/quandl/quandl-python) API for the money market rates, which extracts the data from the [Federal Reserve (FRED)](https://fred.stlouisfed.org).
